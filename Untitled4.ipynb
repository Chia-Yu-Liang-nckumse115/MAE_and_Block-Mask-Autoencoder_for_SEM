```python
import torch

def patchify_imgs(imgs, p):
    """
    imgs: (B, C, H, W)，H=W，且可以被 p 整除
    回傳: (B, L, p*p*C)，L = (H/p) * (W/p)
    """
    B, C, H, W = imgs.shape
    assert H == W and H % p == 0, "影像尺寸必須方形且能被 patch_size 整除"

    h = H // p
    w = W // p

    # (B, C, H, W) → (B, C, h, p, w, p)
    x = imgs.reshape(B, C, h, p, w, p)
    # → (B, h, w, p, p, C)
    x = x.permute(0, 2, 4, 3, 5, 1)
    # → (B, h*w, p*p*C)
    x = x.reshape(B, h * w, p * p * C)
    return x

def unpatchify_imgs(patches, p, C):
    """
    patches: (B, L, p*p*C)
    回傳: imgs: (B, C, H, W)
    """
    B, L, patch_dim = patches.shape
    h = w = int(L ** 0.5)
    assert h * w == L, "patch 數量不是平方數，無法還原成方形圖"

    # (B, L, p*p*C) → (B, h, w, p, p, C)
    x = patches.reshape(B, h, w, p, p, C)
    # → (B, C, h, p, w, p)
    x = x.permute(0, 5, 1, 3, 2, 4)
    # → (B, C, H, W)
    imgs = x.reshape(B, C, h * p, w * p)
    return imgs
```

Patchify / Unpatchify

MAE都不能直接「一次吃整張圖片」，它習慣把圖片拆成固定大小的小區塊（通常 16×16、32×32）。



Patchify 的目的

1. 把大影像變成 token（向量）

2. 讓 Transformer 能處理影像像處理文字一樣

3. 每個 patch 都可以被獨立遮住（mask）來做 MAE 訓練

Unpatchify: 把小 patch 再拼回原本的大影像

1. 接收一堆 patch

2. 按照原本的行列位置排列

3. 拼成完整高×寬的影像

```python
def patchify(x, patch_size=PATCH_SIZE):
    """
    x: (B, C=1, H, W)
    return: (B, N_patches, patch_vec_dim = ps*ps*C)
    """
    B, C, H, W = x.shape
    assert H % patch_size == 0 and W % patch_size == 0, "IMG_SIZE 必須被 PATCH_SIZE 整除"
    h = H // patch_size
    w = W // patch_size
    # (B, C, h, ps, w, ps)
    x = x.reshape(B, C, h, patch_size, w, patch_size)
    # (B, h, w, ps, ps, C)
    x = x.permute(0, 2, 4, 3, 5, 1)
    # (B, h*w, ps*ps*C)
    patches = x.reshape(B, h * w, patch_size * patch_size * C)
    return patches


def unpatchify(patches, patch_size=PATCH_SIZE, in_chans=IN_CHANS):
    """
    patches: (B, N_patches, patch_vec_dim)
    return: (B, C=1, H, W)
    """
    B, N, L = patches.shape
    h = w = int(math.sqrt(N))
    x = patches.reshape(B, h, w, patch_size, patch_size, in_chans)
    # (B, C, H, W)
    x = x.permute(0, 5, 1, 3, 2, 4).reshape(
        B, in_chans, h * patch_size, w * patch_size
    )
    return x
```

Patch Embedding

```python
class PatchEmbed(nn.Module):
    """
    Conv2d with kernel=stride=PATCH_SIZE → 每個 patch 成一個 token
    """
    def __init__(self, img_size=IMG_SIZE, patch_size=PATCH_SIZE, in_chans=IN_CHANS, embed_dim=256):
        super().__init__()
        self.img_size = img_size
        self.patch_size = patch_size
        self.grid_size = img_size // patch_size
        self.num_patches = self.grid_size ** 2
        self.proj = nn.Conv2d(in_chans, embed_dim, kernel_size=patch_size, stride=patch_size)

    def forward(self, x):
        # x: (B, 1, H, W)
        x = self.proj(x)                      # (B, embed_dim, H/ps, W/ps)
        x = x.flatten(2).transpose(1, 2)      # (B, N_patches, embed_dim)
        return x
```

 MAE 本人

```python
class MAE(nn.Module):
    def __init__(
        self,
        img_size=IMG_SIZE,
        patch_size=PATCH_SIZE,
        in_chans=IN_CHANS,
        embed_dim=256,
        encoder_depth=4,
        encoder_heads=8,
        decoder_dim=128,
        decoder_depth=2,
        decoder_heads=4,
        mask_ratio=MASK_RATIO,
    ):
        super().__init__()

        # Patch embedding
        self.patch_embed = PatchEmbed(img_size, patch_size, in_chans, embed_dim)
        num_patches = self.patch_embed.num_patches

        self.mask_ratio = mask_ratio
        self.num_patches = num_patches
        self.patch_size = patch_size
        self.in_chans = in_chans

        # Encoder
        self.pos_embed = nn.Parameter(torch.zeros(1, num_patches, embed_dim))
        enc_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=encoder_heads, batch_first=True)
        self.encoder = nn.TransformerEncoder(enc_layer, num_layers=encoder_depth)

        # Decoder
        self.mask_token = nn.Parameter(torch.zeros(1, 1, decoder_dim))
        self.decoder_embed = nn.Linear(embed_dim, decoder_dim)
        self.decoder_pos_embed = nn.Parameter(torch.zeros(1, num_patches, decoder_dim))
        dec_layer = nn.TransformerEncoderLayer(d_model=decoder_dim, nhead=decoder_heads, batch_first=True)
        self.decoder = nn.TransformerEncoder(dec_layer, num_layers=decoder_depth)
        self.decoder_pred = nn.Linear(decoder_dim, patch_size * patch_size * 1)

        # init
        nn.init.trunc_normal_(self.pos_embed, std=0.02)
        nn.init.trunc_normal_(self.decoder_pos_embed, std=0.02)
        nn.init.trunc_normal_(self.mask_token, std=0.02)

    def random_masking(self, x):
        """
        x: (B, N, C)
        return: x_masked, mask(1=masked), ids_restore, ids_keep
        """
        B, N, _ = x.shape
        len_keep = int(N * (1 - self.mask_ratio))

        noise = torch.rand(B, N, device=x.device)               # each sample a random noise
        ids_shuffle = torch.argsort(noise, dim=1)               # ascend
        ids_restore = torch.argsort(ids_shuffle, dim=1)

        ids_keep = ids_shuffle[:, :len_keep]
        x_masked = torch.gather(x, 1, ids_keep.unsqueeze(-1).repeat(1, 1, x.size(-1)))

        # 1=masked, 0=keep
        mask = torch.ones(B, N, device=x.device)
        mask[:, :len_keep] = 0
        mask = torch.gather(mask, 1, ids_restore)
        return x_masked, mask, ids_restore, ids_keep

    def forward(self, x):
        B = x.size(0)

        # 1) Patch embedding + pos
        x_tokens = self.patch_embed(x)                 # (B, N, C)
        x_tokens = x_tokens + self.pos_embed

        # 2) Random masking
        x_vis, mask, ids_restore, ids_keep = self.random_masking(x_tokens)

        # 3) Encoder on visible tokens
        latent = self.encoder(x_vis)                   # (B, N_keep, C)

        # 4) Prepare tokens for decoder
        dec_tokens = self.decoder_embed(latent)        # (B, N_keep, dec_dim)
        N = self.num_patches
        len_keep = latent.shape[1]

        mask_tokens = self.mask_token.repeat(B, N - len_keep, 1)
        tokens = torch.cat([dec_tokens, mask_tokens], dim=1)     # (B, N, dec_dim)
        tokens = torch.gather(tokens, 1, ids_restore.unsqueeze(-1).repeat(1, 1, dec_tokens.size(-1)))
        tokens = tokens + self.decoder_pos_embed

        # 5) Decoder
        decoded = self.decoder(tokens)                 # (B, N, dec_dim)

        # 6) Predict pixel values for each patch
        pred = self.decoder_pred(decoded)              # (B, N, ps*ps*C)

        # 7) Target patches from original image
        target = patchify(x, patch_size=self.patch_size)  # (B, N, L)

        # 8) MSE on masked patches only
        loss = (pred - target) ** 2
        loss = (loss.mean(dim=-1) * mask).sum() / mask.sum()

        return loss, pred, mask
```

Build model & optimizer

```python
model = MAE(
    img_size=IMG_SIZE,
    patch_size=PATCH_SIZE,
    in_chans=IN_CHANS,
    embed_dim=256,
    encoder_depth=4,
    encoder_heads=8,
    decoder_dim=128,
    decoder_depth=2,
    decoder_heads=4,
    mask_ratio=MASK_RATIO
).to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WD)
```

訓練模型

```python
def train(model, loader, epochs=EPOCHS):
    model.train()
    for ep in range(epochs):
        total = 0.0
        for step, (imgs, _) in enumerate(loader):
            imgs = imgs.to(device)  # (B, 1, H, W)
            loss, _, _ = model(imgs)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total += loss.item()
            if (step + 1) % 20 == 0:
                print(f"Epoch {ep+1} Step {step+1} Loss: {loss.item():.4f}")
        print(f"=== Epoch {ep+1} Avg Loss: {total/len(loader):.4f}")

try:
    train(model, train_loader, epochs=EPOCHS)
except RuntimeError as e:
    print("\n[Hint] 如果遇到 CUDA OOM：")
    print("  1) 把 BATCH_SIZE 調小 (例如 16 或 8)")
    print("  2) IMG_SIZE 改小 (例如 64)")
    print("  3) encoder_depth/decoder_depth 調小")
    raise e
```

換一個寫法

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SEMBlockMaskAE(nn.Module):
    def __init__(self, in_ch=1):
        super().__init__()
        # Encoder
        self.enc = nn.Sequential(
            nn.Conv2d(in_ch, 32, 3, padding=1),   # 256→256
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 256→128

            nn.Conv2d(32, 64, 3, padding=1),      # 128→128
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 128→64

            nn.Conv2d(64, 128, 3, padding=1),     # 64→64
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 64→32

            nn.Conv2d(128, 256, 3, padding=1),    # 32→32
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 32→16
        )  # latent: (B,256,16,16)

        # Decoder
        self.dec = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 2, stride=2),  # 16→32
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(128, 64, 2, stride=2),   # 32→64
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(64, 32, 2, stride=2),    # 64→128
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(32, in_ch, 2, stride=2), # 128→256
            # 最後不用 activation，讓網路自己學輸出範圍
        )

    def forward(self, x_corrupt):
        z = self.enc(x_corrupt)
        recon = self.dec(z)
        return recon
```

```python
import torch

def make_block_mask(x, patch_size=16, mask_ratio=0.5):
    """
    x: (B,1,H,W)，H,W 必須是 patch_size 的整數倍（你是 256 沒問題）
    回傳:
        x_corrupt: 被遮過的影像
        mask_img:  (B,1,H,W)，1 表示該 pixel 被遮掉
    """
    B, C, H, W = x.shape
    assert H % patch_size == 0 and W % patch_size == 0
    h = H // patch_size
    w = W // patch_size
    N = h * w

    device = x.device
    mask = torch.zeros(B, N, device=device)

    num_mask = int(mask_ratio * N)
    for i in range(B):
        idx = torch.randperm(N, device=device)[:num_mask]
        mask[i, idx] = 1.0

    # 展開成 pixel mask
    mask = mask.view(B, 1, h, w)                    # (B,1,h,w)
    mask = mask.repeat(1, 1, patch_size, patch_size)  # (B,1,H,W)

    x_corrupt = x * (1 - mask)  # 被遮掉的地方變 0
    return x_corrupt, mask
```

```python
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SEMBlockMaskAE(in_ch=1).to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

num_epochs = 20
patch_size = 16      # block 大小
mask_ratio = 0.2     # 你指定的 0.2

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for imgs, _ in train_loader:
        imgs = imgs.to(device)  # (B,1,256,256)

        # 做 block mask
        x_corrupt, mask_img = make_block_mask(imgs, patch_size=patch_size,
                                              mask_ratio=mask_ratio)

        # AE 重建
        recon = model(x_corrupt)

        # 對「整張圖」算 MSE（不是只 masked 區域）
        loss = criterion(recon, imgs)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * imgs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch [{epoch+1}/{num_epochs}]  loss = {epoch_loss:.4f}")
```

```python
class SEMBlockMaskAE(nn.Module):
    def __init__(self, in_ch=1):
        super().__init__()
        # Encoder
        self.enc = nn.Sequential(
            nn.Conv2d(in_ch, 32, 3, padding=1),   # 256 → 256
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 256 → 128

            nn.Conv2d(32, 64, 3, padding=1),      # 128 → 128
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 128 → 64

            nn.Conv2d(64, 128, 3, padding=1),     # 64 → 64
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 64 → 32

            nn.Conv2d(128, 256, 3, padding=1),    # 32 → 32
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2),                      # 32 → 16
        )  # latent: (B, 256, 16, 16)

        # Decoder
        self.dec = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 2, stride=2),  # 16 → 32
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(128, 64, 2, stride=2),   # 32 → 64
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(64, 32, 2, stride=2),    # 64 → 128
            nn.ReLU(inplace=True),

            nn.ConvTranspose2d(32, in_ch, 2, stride=2), # 128 → 256
            # 最後不加 Activation，讓網路自由學 output 範圍
        )

    def encode(self, x):
        return self.enc(x)

    def decode(self, z):
        return self.dec(z)

    def forward(self, x_corrupt):
        z = self.encode(x_corrupt)
        recon = self.decode(z)
        return recon
```

```python
def make_block_mask(x, patch_size=16, mask_ratio=0.5):
    """
    x: (B, 1, H, W)；H,W 要能被 patch_size 整除（你 256 沒問題）
    回傳:
      x_corrupt: 被遮掉某些 block 的輸入圖
      mask_img:  (B, 1, H, W)，1 表示該 pixel 被遮掉
    """
    B, C, H, W = x.shape
    assert H % patch_size == 0 and W % patch_size == 0
    h = H // patch_size
    w = W // patch_size
    N = h * w

    device = x.device
    mask = torch.zeros(B, N, device=device)

    num_mask = int(mask_ratio * N)
    for i in range(B):
        idx = torch.randperm(N, device=device)[:num_mask]
        mask[i, idx] = 1.0

    # 拉成 pixel mask
    mask = mask.view(B, 1, h, w)                    # (B,1,h,w)
    mask = mask.repeat(1, 1, patch_size, patch_size)  # (B,1,H,W)

    x_corrupt = x * (1 - mask)  # 被遮掉的地方設為 0
    return x_corrupt, mask
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("device =", device)

model = SEMBlockMaskAE(in_ch=1).to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

num_epochs  = 20
patch_size  = 16          # block 大小（可以改 8 或 32）
mask_ratio  = 0.2         # 你希望的遮罩比例

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for imgs, _ in train_loader:       # 假設 dataloader 已經準備好
        imgs = imgs.to(device)       # (B,1,256,256)，值在 [0,1]

        # 做 block mask 當 masked input
        x_corrupt, mask_img = make_block_mask(
            imgs, patch_size=patch_size, mask_ratio=mask_ratio
        )

        # Autoencoder 重建
        recon = model(x_corrupt)

        # 對**整張圖**算 MSE（不是只有 masked 區）
        loss = criterion(recon, imgs)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * imgs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch [{epoch+1}/{num_epochs}]  loss = {epoch_loss:.5f}")
```

```python
model.eval()
with torch.no_grad():
    imgs, _ = next(iter(train_loader))
    imgs = imgs[:4].to(device)   # 看前 4 張

    x_corrupt, mask_img = make_block_mask(
        imgs, patch_size=patch_size, mask_ratio=mask_ratio
    )
    recon = model(x_corrupt)

imgs_np    = imgs.cpu()
corrupt_np = x_corrupt.cpu()
recon_np   = recon.cpu()

B = imgs_np.size(0)
plt.figure(figsize=(9, 3*B))
for i in range(B):
    # original
    plt.subplot(B, 3, 3*i+1)
    plt.imshow(imgs_np[i,0], cmap='gray')
    plt.title("original")
    plt.axis("off")

    # masked input
    plt.subplot(B, 3, 3*i+2)
    plt.imshow(corrupt_np[i,0], cmap='gray')
    plt.title("masked input")
    plt.axis("off")

    # recon
    plt.subplot(B, 3, 3*i+3)
    plt.imshow(recon_np[i,0], cmap='gray')
    plt.title("recon")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
